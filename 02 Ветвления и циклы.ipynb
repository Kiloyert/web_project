## Задание 1: Чётное или нечётное
Напишите программу, которая запрашивает у пользователя целое число и выводит, является ли оно чётным или нечётным.

````python
a=int(input())
if a % 2 == 1:
  print(f'Число {a} нечетное')
else:
  print(f'Число {a} четное')
````

## Задание 2: Квадратное уравнение (простой случай)
Программа получает коэффициенты a, b, c для уравнения a*x² + b*x + c = 0. Вычислите дискриминант. Если он положительный, выведите сообщение "Два корня", если равен нулю — "Один корень", если отрицательный — "Действительных корней нет". Сами корни находить не нужно.

````python
a, b, c = input().split()
a, b, c= int(a), int(b), int(c)
D=b**2-4*a*c
if D>0 :
  print('Два корня')
elif D==0 :
  print('Один корень')
else:
  print("Действительных корней нет")
````

## Задание 3: Максимум из трёх
Запросите у пользователя три числа и выведите наибольшее из них, используя только ветвления (без функций max и сортировки).

````python
a, b, c = map(int, input().split())
if a>=b and a>=c :
  print(f'Наибольшее число {a}')
elif b>=a and b>=c :
  print(f'Наибольшее число {b}')
elif c>=a and c>=b :
  print(f'Наибольшее число {c}')
````

## Задание 4: Високосный год
Год является високосным, если он кратен 4, но не кратен 100, либо кратен 400. Напишите программу, которая определяет по введённому году, високосный он или нет.

````python
year= int(input())
if (year%4 == 0 and year % 100 != 0) or year % 400 == 0:
  print('Год високосный')
else:
  print('Год не високосный')
````

## Задание 5: Сумма чисел от 1 до N
Запросите у пользователя число N (натуральное). Вычислите и выведите сумму всех целых чисел от 1 до N.

````python
n=int(input())
cnt=0
sum=0
while cnt != n:
  cnt += 1
  sum +=cnt
print(sum)
````

## Задание 8: Сумма цифр
Запросите у пользователя целое положительное число. Найдите сумму его цифр, используя цикл while.

````python
num=int(input())
sum = 0
cnt = num

while cnt > 0:
    sum += cnt % 10
    cnt //= 10
print(sum)
````

## Задание 9: Обратный отсчёт
Запросите число N. Затем с помощью цикла while выведите числа от N до 1, а в конце слово "Старт!".

````python
n=int(input())
cnt=n
while cnt!=0:
  print(cnt)
  cnt -=1
print("Старт!")
````

## Задание 12: Числа Фибоначчи
Выведите N первых чисел последовательности Фибоначчи. Напоминание: F₀=0, F₁=1, Fₙ = Fₙ₋₁ + Fₙ₋₂.

````python
n=int(input()) #считаем что n всегда больше 2
F_0 = 0
F_1=1
cnt=2
print(F_0)
print(F_1)
while cnt!=n:
  ans=F_0 + F_1
  F_0=F_1
  F_1=ans
  cnt +=1
  print(ans)
````

## Задание 13: Калькулятор с обработкой ошибок

````python
a,b = map(int, input().split())
oper=input()
if oper == '+':
  print(f'Сумма чисел {a} и {b} равна {a+b}')
elif oper == '-':
  print(f'Разность чисел {a} и {b} равна  {a-b}')
elif oper == '*':
  print(f'Произведение чисел {a} и {b} равно {a*b}')
elif oper == '/':
  if b==0:
    print("На ноль делить нельзя")
  else:
    print(a/b)
````

## Задание 14: Угадай число (компьютер загадывает)
Программа "загадывает" число от 1 до 100 (используйте фиксированное, например, 42). Пользователь пытается угадать. На каждый ход программа отвечает "Больше", "Меньше" или "Угадал!". Подсчитайте количество попыток.

````python
import random
random_number = random.randint(1, 100)
my_number=int(input())
while my_number != random_number:
  if my_number < random_number:
    print('Ваше число меньше чем загаданное')
  elif my_number > random_number:
    print('Ваше число больше чем загаданное')
  my_number=int(input())
print('Угадал')
````
